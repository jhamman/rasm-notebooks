```python
import numpy as np
from netCDF4 import Dataset,num2date, date2num,date2index
import pandas as pd
import datetime
from mpl_toolkits.basemap import Basemap, cm
from glob import glob
import os
from matplotlib.ticker import FuncFormatter
```

```python
outdir='/usr1/jhamman/Dropbox/UW_RASM_shared/2013_Seattle_Meeting/plots/'
gridFile = '/raid/jhamman/RASM_masks/domain.lnd.wr50a_ar9v4.100920.nc'
runs={'r35RB1a':'/raid/jhamman/RASM_results/r35RB1a/means/r35RB1a.vic.ha.1990-1999-??.monthly.mean.nc',
      'S17_r35RB1a':'/raid/jhamman/RASM_results/Snow17/means/r35RB1a.snow17.1990-1999-??.mean.nc',
      'S17_Sheffield':'/raid/jhamman/RASM_results/Snow17/means/Sheffield.1990-1999-??.mean.nc'}

# Get areas and land fractions from grid file
re = 6.37122e6

f = Dataset(gridFile,'r')
frac = f.variables['frac'][:]
area = f.variables['area'][:]*re*re/1000/10000*frac # area in square kilometers
lats = f.variables['yc'][:,:]
total_area = np.sum(area)
f.close()
```

```python
# Helper functions:
def make_datetimes(date_list):
    datetime_list=[]
    for i in date_list:
        datetime_list.append(datetime.datetime(i.year,i.month,i.day,i.hour,i.minute,i.second))
    return datetime_list

def find_doy(date):
    doy = date.timetuple().tm_yday
    return doy

def date_from_string(s):
    date = datetime.datetime(int(s[:4]),int(s[4:6]),int(s[6:8]))
    return date
    
def get_nsidc_dates(inpath):
    files = glob(inpath)
    files.sort()
    dates = []
    for fi in files:
        fn = os.path.basename(fi)
        date1 = date_from_string(fn[2:10])
        date2 = date_from_string(fn[11:19])
        dates.append(date1+(date2-date1)/2)
    return files, dates

def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1f' % (x*1e-6)
formatter = FuncFormatter(millions)
```

```python
# get a timeseries of NSIDC extents
files, dates = get_nsidc_dates("/raid/jhamman/data/snow/NL19*")
temp = np.zeros(len(files))
for i, fi in enumerate(files):
    f = Dataset(fi,'r')    
    snowice = f.variables['snowice'][:]
    f.close()
    snowice[snowice >1] = 0
    snowice[lats<clipping_lat] = 0
    snowice[snowice !=0] = 1
    temp[i] = (snowice*area).sum()

nsidc = pd.Series(temp,index=dates)
    
```

```python
# time series plots of snow covered extent
tsnow = {}
for i, (run,infile) in enumerate(runs.iteritems()):
    print infile
    f=Dataset(infile, 'r')
    snowice = f.variables['Swq'][:]
    lats = f.variables['latitude'][:]
    time=make_datetimes(num2date(f.variables['time'][:],units=f.variables['time'].units,calendar=f.variables['time'].calendar))
    f.close()
    print 'lats:', lats.shape
    print 'snow:', snowice.shape
    snowice[snowice < 10] = 0
    snowice[:,lats<clipping_lat] = 0
    snowice[snowice !=0] = 1
    tsnow[run] = pd.Series((snowice*area).sum(axis=2).sum(axis=1),index=time)

snow_cover = pd.DataFrame(tsnow)
```
