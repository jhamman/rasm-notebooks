```python
import os, sys
import numpy as np
import pandas as pd
import datetime
import matplotlib as mpl
```

```python
def date_list(years,months,days,hours):
    dates = []

    for i in xrange(len(years)):
        dates.append(datetime.datetime(years[i],months[i],days[i],hours[i]))
        
    return dates
```

```python
inPath1 = "/raid/jhamman/RASM_results/offline_VIC/Mar_18/"
fluxFile1 = "fluxes_60.030_-108.898"
snowFile1 = "snow_60.030_-108.898"

sData1= np.genfromtxt(os.path.join(inPath1,snowFile1))
fData1 = np.genfromtxt(os.path.join(inPath1,fluxFile1))

sYear1 = sData1[:,0].astype(int)
sMonth1 = sData1[:,1].astype(int)
sDay1 = sData1[:,2].astype(int)
sHour1 = sData1[:,3].astype(int)

dates1 = date_list(sYear1,sMonth1,sDay1,sHour1)
dfSnow1 = pd.DataFrame(sData1, index = dates1,columns=['Year', 'Month', 'Day', 'Hour', 'SWE', 'Snow_Depth',
'Snow_Canopy', 'Snow_Cover','Snow_Surf_Temp','Snow_Albedo'])
dfFlux1 = pd.DataFrame(fData1, index = dates1,columns=['Year', 'Month', 'Day', 'Hour', 'Precipitation',
'Evaporation','Net_Shortwave','Surface_Temp','Albedo','Air_Temp','In_Shortwave'])
```

```python
inPath2 = "/raid/jhamman/RASM_results/offline_VIC/new_bonly_one/"
fluxFile2 = "fluxes_60.030_-108.898"
snowFile2 = "snow_60.030_-108.898"

sData2= np.genfromtxt(os.path.join(inPath2,snowFile2))
fData2 = np.genfromtxt(os.path.join(inPath2,fluxFile2))

sYear2 = sData2[:,0].astype(int)
sMonth2 = sData2[:,1].astype(int)
sDay2 = sData2[:,2].astype(int)
sHour2 = sData2[:,3].astype(int)

dates2 = date_list(sYear2,sMonth2,sDay2,sHour2)
dfSnow2 = pd.DataFrame(sData2, index = dates2,columns=['Year', 'Month', 'Day', 'Hour', 'SWE', 'Snow_Depth', 'Snow_Canopy', 'Snow_Cover','Snow_Surf_Temp','Snow_Albedo'])
dfFlux2 = pd.DataFrame(fData2, index = dates2,columns=['Year', 'Month', 'Day', 'Hour', 'Precipitation',
'Evaporation','Net_Shortwave','Surface_Temp','Albedo','Air_Temp','In_Shortwave'])
```

```python
def water_day(indate):
    doy = indate.timetuple().tm_yday
    if doy >= 274:
        outdate = doy - 274
    else:
        outdate = doy + 91
    return outdate
```

```python
plt.figure()

dfSnow1.SWE.groupby(lambda x:water_day(x)).mean().plot(style='k--', label='Original'); 
dfSnow2.SWE.groupby(lambda x:water_day(x)).mean().plot(label='Modified')
plt.legend(loc='best')
xlabel('Day of Water Year')
ylabel('Snow Water Equivalent (mm)')
plt.xlim((0,366))
```
