```python
def make_datetimes(date_list):
    datetime_list=[]
    for i in date_list:
        datetime_list.append(datetime.datetime(i.year,i.month,i.day,i.hour,i.minute,i.second))
    return datetime_list

def find_doy(date):
    doy = date.timetuple().tm_yday
    return doy
```

```python
# Get areas and land fractions from grid file

f = Dataset(gridFile,'r')
frac = f.variables['frac'][:]
area = f.variables['area'][:]*frac*re*re
lons = f.variables['xc'][:]
lats = f.variables['yc'][:]


f.close()
name = 'full_'
lon_min = 0
lon_max = 360
lat_min = -90
lat_max = 90

name = 'NA_'
lon_min = 190
lon_max = 340
lat_min = 45
lat_max = 90

name = 'RU_'
lon_min = 30
lon_max = 190
lat_min = 45
lat_max = 90

mask = np.ma.masked_inside(lons,lon_min,lon_max).mask*np.ma.masked_inside(lats,lat_min,lat_max).mask
frac *= mask
total_area = np.sum(area*frac)
```
