```python
import numpy as np
from netCDF4 import Dataset,num2date, date2num,date2index
from scipy.spatial import cKDTree
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.basemap import Basemap, cm
import pandas as pd
```

```python
# Helper functions:
def make_datetimes(date_list):
    datetime_list=[]
    for i in date_list:
        datetime_list.append(datetime.datetime(i.year,i.month,i.day,i.hour,i.minute,i.second))
    return datetime_list
```

```python
stFile = "/usr1/jhamman/Dropbox/RASM_Joe/python_scripts/Streamflow_Routing/convolution/python/r35RB1a_1989-1999_Streamflow_All.nc"
f = Dataset(stFile, 'r')
lats = f.variables['latitudes'][:]
lons = f.variables['longitudes'][:]
sflow=f.variables['Streamflow'][:]
time=make_datetimes(num2date(f.variables['time'][:],units=f.variables['time'].units,calendar=f.variables['time'].calendar))
f.close()
drFile = "/raid/jhamman/RASM_masks/RASM_Drainage_Mask.nc"
f = Dataset(drFile,'r')
dmask = f.variables['fraction'][:]
xc = f.variables['xc'][:]
yc = f.variables['yc'][:]
f.close()
dmask[dmask>0]=1
dmask = np.ma.masked_where(dmask<=0,dmask)

temp = {}
for i in xrange(sflow.shape[1]):
    temp[i]=pd.Series(sflow[:,i],index=time)
    
sflow_df = pd.DataFrame(temp)
```
