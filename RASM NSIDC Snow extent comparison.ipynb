##Read NSIDC data##

Data are stored in flat binary files with arrays of 721 columns by 721 rows of 1­-byte unsigned integers. Each data file contains one array.

Read a sample file (`NL19900101-19900107.v03.SI`) into a nympy array named `snowice`:

```python
import numpy as np

infile = '/Users/nijssen/junk/nsidc_weekly_snow_seaice/data/NL19900101-19900107.v03.SI'
nx = ny = 721
with open(infile, 'rb') as f:
    rawdata = f.read(nx * ny)

snowice = np.frombuffer(rawdata, dtype=np.uint8)
snowice.resize([nx, ny]);
snowice.shape
```

```python
import matplotlib as mpl
import itertools

cmap = mpl.colors.ListedColormap(['gray'] * 256)
vals = (0,1,2,3,4,5,253,254,255)
colors = ('green', 'white', 'cyan', 'cyan', 'blue', 'white', 'red', 'gray', 'blue')
for val, color in itertools.izip(vals, colors):
    cmap.colors[val] = color  
imgplot = plt.imshow(snowice)
imgplot.set_cmap(cmap)
plt.colorbar();
```

```python
from netCDF4 import Dataset
ncout = Dataset('/Users/nijssen/junk/easegrid.nc', 'w', format='NETCDF4')
ni = ncout.createDimension('ni', 721)
nj = ncout.createDimension('nj', 721)
outlon = ncout.createVariable('xc', 'f4', ('nj', 'ni',), fill_value=lon[0,0])
outlat = ncout.createVariable('yc','f4',('nj', 'ni',), fill_value=lat[0,0])
outmask = ncout.createVariable('mask','u1',('nj', 'ni',), fill_value=mask[0,0])
outsnow = ncout.createVariable('snowice','u1',('nj', 'ni',), fill_value=snowice[0,0])
outlon.long_name = 'longitude of grid cell center'
outlat.long_name = 'latitude of grid cell center'
outmask.long_name = 'domain mask'
outsnow.long_name = 'Snow ice status'
outlon.units = 'degrees_east'
outlat.units = 'degrees_north'
outsnow.units = '-'
outmask.units = '-'
outsnow.coordinates = "xc yc"
outmask.coordinates = "xc yc"
outlon[:,:] = lon
outlat[:,:] = lat
outmask[:,:] = mask
outsnow[:,:] = snowice
ncout.close()
```
