```python
import numpy as np
from netCDF4 import Dataset,num2date, date2num,date2index
import pandas as pd
import datetime
from mpl_toolkits.basemap import Basemap, cm

re = 6.37122e6
```

```python
outdir='/usr1/jhamman/Dropbox/UW_RASM_shared/2013_Seattle_Meeting/plots/'

gridFile = '/raid/jhamman/RASM_masks/domain.lnd.wr50a_ar9v4.100920.nc'
rasmFiles={
'r30RB1g':'/raid/jhamman/RASM_results/r30RB1g/r30RB1g.vic.ha.198909-199912.daily.nc',
'r33RBVIC60':'/raid/jhamman/RASM_results/r33RBVIC60/r33RBVIC60.vic.ha.198909-199912.daily.nc',
'r33RBVIC70':'/raid/jhamman/RASM_results/r33RBVIC70/r33RBVIC70.vic.ha.1990-1999.daily.nc',
'r35RB1a':'/raid/jhamman/RASM_results/r35RB1a/r35RB1a.vic.ha.198909-199912.daily.nc'}

runcolors={
'r30RB1g':'r',
'r33RBVIC60':'b',
'r33RBVIC70':'g',
'r35RB1a':'c'}

start_datetime = datetime.datetime(1990,1,1)
end_datetime = datetime.datetime(1999,12,31)
```

```python
def make_datetimes(date_list):
    datetime_list=[]
    for i in date_list:
        datetime_list.append(datetime.datetime(i.year,i.month,i.day,i.hour,i.minute,i.second))
    return datetime_list

def find_doy(date):
    doy = date.timetuple().tm_yday
    return doy
```

```python
xmin = 140; xmax=210
ymin = 40; ymax=90

# Get areas and land fractions from grid file

f = Dataset(gridFile,'r')
frac = f.variables['frac'][ymin:ymax,xmin:xmax]
area = f.variables['area'][ymin:ymax,xmin:xmax]*frac*re*re
total_area = np.sum(area)

f.close()
pcolor(area)
```

```python
# Radiation Balance

variables = ['Latht','Senht','Grdht','Swin','Swnet','Lwin','Lwnet','Netrad']
data_frames = {}
for run,dfn in rasmFiles.iteritems():
    rad_vars = {}
    temp = {}
    time_vars = {}
    f=Dataset(dfn,'r')
    
    for var in variables:
        data = f.variables[var][:,ymin:ymax,xmin:xmax]
        rad_vars[var]=(data*area).sum(axis=2).sum(axis=1)/total_area
        
        
    time=make_datetimes(num2date(f.variables['time'][:],units=f.variables['time'].units,calendar=f.variables['time'].calendar))
    f.close()

    data_frames[run]= pd.DataFrame(rad_vars, index=time)
```
