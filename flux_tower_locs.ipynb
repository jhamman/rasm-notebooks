```python
gridFile = "/usr1/jhamman/RASM/VIC/soil_param/params.nc"

f = Dataset(gridFile,'r')
xc = f.variables['xc'][:]
yc = f.variables['yc'][:]
Cv = f.variables['Cv'][:]

if xc.min()>=0:
    posinds = np.nonzero(xc>180)
    xc[posinds] -= 360
    print 'adjusted xc lon minimum'
```
