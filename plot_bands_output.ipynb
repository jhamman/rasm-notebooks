```python
import os, sys
import numpy as np
import pandas as pd
import datetime
import matplotlib as mpl
```

```python
def date_list(years,months,days):
    dates = []

    for i in xrange(len(years)):
        dates.append(datetime.datetime(years[i],months[i],days[i]))
        
    return dates

def water_day(indate):
    doy = indate.timetuple().tm_yday
    if doy >= 274:
        outdate = doy - 274
    else:
        outdate = doy + 91
    return outdate

```

```python
inPath1 = "/raid/jhamman/alaska/outputs/"
dataFile = "snow_bands_60.319_-139.604"
bandsFile = "snow_bands_60.319_-139.604"

outdata= np.genfromtxt(os.path.join(inPath1,dataFile))
banddata = np.genfromtxt(os.path.join(inPath1,bandsFile))

sYear1 = outdata[:,0].astype(int)
sMonth1 = outdata[:,1].astype(int)
sDay1 = outdata[:,2].astype(int)

dates1 = date_list(sYear1,sMonth1,sDay1)
#data = pd.DataFrame(outdata, index = dates1,columns=['Year', 'Month', 'Day', 'SWE', 'Snow_Depth',
#'Snow_Canopy', 'Snow_Cover','Snow_Surf_Temp','Snow_Albedo'])
bands = pd.DataFrame(banddata, index = dates1,columns=['Year', 'Month', 'Day','Albedo1','Albedo2','Albedo3','Albedo4','Albedo5',
'SnowCover1','SnowCover2','SnowCover3','SnowCover4','SnowCover5','SnowDepth1','SnowDepth2','SnowDepth3','SnowDepth4','SnowDepth5',
'SnowMelt1','SnowMelt2','SnowMelt3','SnowMelt4','SnowMelt5','SnowPackT1','SnowPackT2','SnowPackT3','SnowPackT4','SnowPackT5',
'SnowSurfT1','SnowSurfT2','SnowSurfT3','SnowSurfT4','SnowSurfT5','SnowSWEBand1','SnowSWEBand2','SnowSWEBand3','SnowSWEBand4','SnowSWEBand5'])
```

```python
bands.SnowSWEBand5.resample('M',how='mean').plot()
#bands.SnowSWEBand5.resample('M',how='mean').plot(c='r')
#bands.SnowSWEBand4.resample('M',how='mean').plot(c='g')
```

```python
bands.SnowSWEBand1.groupby(lambda x:water_day(x)).mean().plot(style='k', label='1');
bands.SnowSWEBand2.groupby(lambda x:water_day(x)).mean().plot(style='r', label='2'); 
bands.SnowSWEBand3.groupby(lambda x:water_day(x)).mean().plot(style='b', label='3'); 
bands.SnowSWEBand4.groupby(lambda x:water_day(x)).mean().plot(style='g', label='4'); 
bands.SnowSWEBand5.groupby(lambda x:water_day(x)).mean().plot(style='c', label='5'); 
plt.legend(loc='best')
```
